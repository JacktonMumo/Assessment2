```python
import nltk
```

```python
import requests
from bs4 import BeautifulSoup 
URL = "https://nation.africa/kenya/life-and-style/dn2/men-enjoy-a-genuine-public-display-of-love-4576356"
page = requests.get(URL)
soup = BeautifulSoup(page.content, "html.parser")
div = soup.findAll("div", class_ ="article-content xxx")
soup = BeautifulSoup(str(div))
text = soup.get_text()
print(text)
```

```python
from nltk.tokenize import sent_tokenize, word_tokenize

token_sentences = sent_tokenize(text)
token_sentences
token_words = word_tokenize(text)
token_words
```

```python
from nltk.corpus import stopwords
print(stopwords.fileids())
mystopwords = set(stopwords.words('english'))
mystopwords
```

```python
filtered_words =[]
for word in token_words:
    if word.lower() not in mystopwords:
        filtered_words.append(word)
print("the new words without the common words is:",filtered_words)
```

```python
clean_words = [word for word in filtered_words if word.isalpha()]
clean_words
```

```python
from nltk.probability import FreqDist
frequency = FreqDist(clean_words)
frequency.most_common(30)
```
