```python
import pandas
import sklearn
```

```python
Tweets=pandas.read_csv('chatgpt.csv')
Tweets.head(5)
```

```python
Tweets.isnull().sum()
```

```python
from sklearn import model_selection 
X_train,X_test,Y_train,Y_test =model_selection.train_test_split(Tweets['tweets'],Tweets['labels'],test_size=0.3,random_state=42)

Y_test
```

```python
from sklearn.feature_extraction.text import CountVectorizer 
vectorizer=CountVectorizer(lowercase=True,stop_words='english')
X_train_new=vectorizer.fit_transform(X_train)
X_train_new
```

```python
from sklearn.linear_model import LogisticRegression,SGDRegressor,SGDClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeRegressor

model=LogisticRegression()
model.fit(X_train_new,Y_train)
```

```python
X_test_new=vectorizer.transform(X_test)
prediction=model.predict(X_test_new)
prediction
print(Tweets.groupby('tweets').size())
```
