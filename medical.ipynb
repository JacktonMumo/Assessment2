```python
medic['Diagnosis Category'].fillna('unknown',inplace=True)
medic['Diagnosis Sub Category'].fillna('unknown',inplace=True)
medic['Treatment Category'].fillna('unknown',inplace=True)
medic['Treatment Sub Category'].fillna('unknown',inplace=True)
medic['Age Range'].fillna('unknown',inplace=True)
medic['Patient Gender'].fillna('unknown',inplace=True)
medic['Findings'].fillna('unknown',inplace=True)
medic.isnull().sum()
```
